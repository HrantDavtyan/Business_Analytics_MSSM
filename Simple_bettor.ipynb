# Financial Analytics - Simple bettor

This notebook provides introduces Monte Carlo simulation scenario with simple betting strategy (each time the same bet) described as a Roulette game.

```python
import random
import matplotlib.pyplot as plt
```

Define a function, which will randomly generate an integer from the (1,37) range. If that number is from the range [1,18] then you win, else you loose. Comparing to the Roulette game, [1,18] can be considered as the black patterns, [19,36] as the red, and 37 as the green one, while we are betting on black.

```python
def spinner():
    slot = random.randint(1,37)
    if slot == 37:
        return False
    elif slot <= 18:
        return True
    else:
        return False
```

Now let's use the function above to create a "Simple bettor". The latter always bets the same amount of money (bet) for several periods (periods) given the initial budget s/he has (budget). If s/he wins, the budget increases by the amount of bet, otherwise it decreases. The game stops in one of the two cases: either budget becomes 0 and bettor is not able to bet anymore, or the number of periods the bettor decided to play comes to its end.

We want to plot all this values using matplotlib, which means we should have the X_axis and Y_axis of the plot defined. X_axis will show the periods, while Y_axis will show the budget at the given period. For that purpose we will create an empty list for each axis to be appended with values after each period. 

```python
def simple_bettor(budget,bet,periods):
    X_axis = []
    Y_axis = []

    currentPeriod = 1
    while currentPeriod <= periods:
        if spinner():
            budget += bet
        else:
            budget -= bet
        
        X_axis.append(currentPeriod)
        Y_axis.append(budget)
        currentPeriod += 1
        
    plt.plot(X_axis,Y_axis)
```

Now, given the simple_bettor() function above, let's run it for several times to simulate different possible futures. And this is what Monte Carlo simulation is doing: simulating many different possible futures.

```python
periods = input("Please, input the number of periods: ")
x = 0
while x < 100:
    simple_bettor(10000,100,periods)
    x += 1

plt.ylabel('Budget')
plt.xlabel('Periods')
plt.show()
```
